```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
## book url
book_url = 'https://books.toscrape.com/'
response = requests.get(book_url)
```

```python
print(response)
```

```python
### parsing the html
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
##find all the books 
books = soup.find_all("article", class_='product_pod')
```

```python
print(books)
```

```python
book_data = []
```

```python
for book in books:
    title = book.h3.a['title']
    price = book.find('p', class_='price_color').text
    rate_class  = book.find('p', class_='star-rating')
    availability_class = book.find('p', class_='instock').text
    availability = 1 if ' In stock' in availability_class else 0 
    ## get the rating
    rating = [cls for cls in rate_class.get('class') if cls !='star-rating'][0]
    ## appending into the book_data
    book_data.append({
    'title': title,
    'price': price,
    'rating': rating,
    'availability': availability, 
    })
```

```python
book_data
```

```python
df = pd.DataFrame(book_data)
```
